**Cleaning the text**

```
stop_words = list(nlp.Defaults.stop_words)
lemmatizer = WordNetLemmatizer()
```

**Lemmatization**

```
def lemmatization(iss_str):
  token_list = []
  newString = ''
  for token in nlp(iss_str):
    token_list.append(token.lemma_)
  s = ' '.join(token_list)
  s = re.sub(r'-+[a-zA-Z]+-+', '', s)
  return s
```

```
def cleaner(iss_str):
    iss_str = re.sub(r'[()]', '', iss_str)
    iss_str = re.sub(r'[.,]', '', iss_str)
    iss_str = re.sub(r'\n', ' ', iss_str)
    iss_str = re.sub(r'st\b', '', iss_str)
    iss_str = re.sub(r'–', ' ', iss_str)
    iss_str = re.sub(r'…', '', iss_str)
    iss_str = re.sub(r'[“”]', '', iss_str)
    iss_str = re.sub(r'’s', '', iss_str)
    iss_str = re.sub(r'[;:?]', '', iss_str)
    iss_str = re.sub(r'[-]', ' ', iss_str)
    iss_str = iss_str.lower()
    tokens = iss_str.split()
    newString = lemmatization(iss_str)
    return newString
```

**Removing Stop words**

```
iss['clean_details'] = iss['Detail'].apply(lambda x : cleaner(x))
```

**Entity Recognition**

```
for index, item in iss.iterrows():
  doc = nlp(item[2])
  displacy.render(doc, style = 'ent', jupyter = True)
  print('\n')
  print('------------------------------------------------------------------------------------------------------------------------------------------------------------------------')
  print('\n')
```

**Creating a column for Entities**

```
def show_entity(doc):
  entity = {}
  doc = nlp(doc)
  if doc.ents:
    for ent in doc.ents:
      entity[ent.text] = ent.label_
  return entity
```

```
iss['entity'] = iss['clean_details'].apply(lambda x : show_entity(x))
```
